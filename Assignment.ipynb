```
import pandas as pd
import seaborn as sns
```

```
filepath = 'https://raw.githubusercontent.com/elizabethts/Tuesday-June-Fourth/master/survey.csv'
```

```
survey = pd.read_csv(filepath)
```

```
survey.head()
```

```
sns.distplot(survey.fillna(0).iloc[:, 1]);
```

```
sns.pairplot(survey.fillna(0));
```

```
#pairplot w/ Linear Regression Lines
sns.pairplot(survey, kind='reg',
            plot_kws={'line_kws':{'color':'red'}, 'scatter_kws': {'alpha': 0.1}});
```

```
# Kernel density estimation
sns.jointplot(x='What is your programming level?', y="What is your math level?", data=survey, kind="kde");
```
